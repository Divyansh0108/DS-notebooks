```python
def clean_model_names(model): # function to clean the model names 
    if model.startswith('Rover'):
        return model[len('Rover'):]
    elif model.startswith('Cooper'):
        return model[len('Cooper'):]
    else:
        return model

# Apply the cleaning function to the 'Model' column
df['Model'] = df['Model'].apply(clean_model_names)

# Example to show what happens
print(df['Model'])
```

```python
# Remove 'kmpl' and 'km/kg' from the Mileage column
df['Mileage'] = df['Mileage'].str.replace('kmpl','').str.replace('km/kg','').astype(float)
```

```python
# multiplying the mileage by 1.36 to convert km/kg to kmpl where fuel is CNG
df.loc[df['Fuel_Type'] == 'CNG', 'Mileage'] = df['Mileage']*1.36
```

```python
# removing cc from the Engine column and converting it to float
df['Engine'] = df['Engine'].str.replace('CC','').astype(float)
```

```python
df['Power'].isnull().sum() # check for missing values in the Power column
```

```python
import pandas as pd
import numpy as np

# Replace 'bhp', 'null', and any spaces with NaN
df['Power'] = df['Power'].str.replace('bhp', '', regex=False)  # Remove 'bhp'
df['Power'] = df['Power'].str.strip()  # Strip whitespace
df['Power'] = df['Power'].replace(['null', ''], np.nan)  # Replace 'null' and empty strings with NaN

# Convert the 'Power' column to float
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')  # Safely convert to float, coercing invalid entries to NaN

# Alternatively, replace NaN with the median value if required
median_power = df['Power'].median()
df['Power'] = df['Power'].fillna(median_power)
```

```python
# handling new price column 
df['New_Price'].isnull().sum() # check for missing values in the New_Price column
```

```python
# Remove 'Lakh' and 'Cr' from the New_Price column 
df['New_Price'] = df['New_Price'].str.replace('Lakh','').str.replace('Cr','').astype(float)
```

```python
# Multiply the New_Price by 100000 if the value is in Lakh
df.loc[df['New_Price'] < 100, 'New_Price'] = df['New_Price']*100000
```

```python
# adding units like Lakh to the column name,bhp to Power column and kmpl to Mileage column, and CC to engine column 
df.rename(columns={'New_Price':'New_Price_Lakh', 'Power':'Power_bhp', 'Mileage':'Mileage_kmpl', 'Engine':'Engine_CC'}, inplace=True)
```

### Findings from Used Car Dataset Analysis

- **Year Range**: 
  - **Years range from 1996 to 2019**, indicating that the dataset includes both the latest models and older model cars. This high range in years shows a broad spectrum of car ages in the used car market.

- **Kilometers Driven**:
  - **On average, used cars have been driven about 58,000 kilometers**. However, there's a significant gap between the minimum and maximum values with the **max at 650,000 kilometers**, suggesting the presence of an **outlier**. This record might be considered for removal to normalize the dataset.

- **Mileage**:
  - The **minimum value of mileage is 0**, which is unlikely since cars with zero mileage wouldn't typically be sold as used. This indicates a **possible data entry error**.

- **Engine and Power**:
  - Both **Engine and Power features appear to have outliers** with data distributions being **right-skewed**. This skewness suggests that most cars have lower values for these attributes, with a few cars having significantly higher values.

- **Seats**:
  - The **average number of seats in a car is 5**. The number of seats is an **important feature** that can contribute to the car's price, reflecting its size and utility.

- **Price**:
  - The **maximum price observed for a used car is 160,000**, which seems unusually high for the used car market. This could point to an **outlier or another data entry issue**.

These findings suggest several data cleaning and normalization steps might be necessary before further analysis or modeling.

```python
# Function for log transformation of the column
def log_transform(df, col):
    for colname in col:
        if (df[colname] == 1.0).all():
            df[colname + '_log'] = np.log(df[colname] + 1)
        else:
            df[colname + '_log'] = np.log(df[colname])
    df.info()

# Apply log transformation to the specified columns
log_transform(df, ['Kilometers_Driven', 'Price'])

# Log transformation of the feature 'Kilometers_Driven'
plt.figure(figsize=(8, 6))
sns.distplot(df["Kilometers_Driven_log"], axlabel="Kilometers_Driven_log")
plt.title("Distribution of Kilometers_Driven_log")
plt.show()

# Log transformation of the feature 'Price'
plt.figure(figsize=(8, 6))
sns.distplot(df["Price_log"], axlabel="Price_log")
plt.title("Distribution of Price_log")
plt.show()
```

```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18)) # create a 4x2 grid of subplots
columns_titles = { # dictionary of column names and their titles
    'Location': "Location Vs Price",
    'Transmission': "Transmission Vs Price",
    'Fuel_Type': "Fuel_Type Vs Price",
    'Owner_Type': "Owner_Type Vs Price",
    'Brand': "Brand Vs Price",
    'Model': "Model Vs Price",
    'Seats': "Seats Vs Price",
    'Car_Age': "Car_Age Vs Price"
}

for idx, (col, title) in enumerate(columns_titles.items()): # loop through the columns_titles dictionary
    ax = axarr[idx // 2][idx % 2] # get the current axis
    if col in ['Brand', 'Model']: # plot the top 10 values for Brand and Model
        df.groupby(col)['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=ax, fontsize=12)
    else:
        df.groupby(col)['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=ax, fontsize=12)
    ax.set_title(title, fontsize=18)

plt.subplots_adjust(hspace=1.0, wspace=0.5)
sns.despine() # remove the top and right spines from the plot
```
